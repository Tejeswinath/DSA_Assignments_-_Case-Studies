# 4)  Fill in all the missing values present in all the columns in the dataset

```python
ship_titanic_data.isna().sum()
```

```python
fill_nans=ship_titanic_data['Age']
fill_nans.hist(figsize=(8,8))
```

# 5) Check and handle outliers in at least 3 columns in the dataset

```python
plt.boxplot(ship_titanic_data['Fare'],vert=False)
plt.title('Box Plot of Fare')
```

```python
tenth_percentile1=np.percentile(ship_titanic_data['Fare'],10)
nintieth_percentile1=np.percentile(ship_titanic_data['Fare'], 90)
print(tenth_percentile1,nintieth_percentile1)
a=np.where(ship_titanic_data['Fare']<tenth_percentile1,tenth_percentile1,ship_titanic_data['Fare'])
a=np.where(a>nintieth_percentile1,nintieth_percentile1,a)
print('New Array:',a)
```

```python
plt.boxplot(a, vert=False)
plt.title("Box Plot of Fare")
```

```python
plt.boxplot(ship_titanic_data['Parch'], vert=False)
plt.title('Box Plot of Parch')
```
