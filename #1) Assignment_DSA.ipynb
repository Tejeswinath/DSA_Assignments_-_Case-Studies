```python
#1)  Read the dataset(iris) to python environment.
import pandas as pd
# reading the dataset(iris) in collected data(coll_data)
coll_data=r'C:\Users\Tejes\Downloads\iris.xls'

# using read_excel function to load the data
coll_data=pd.read_excel(coll_data)

# checking the first five index of the dataset 
coll_data.head()
```

```python
#2) Display the columns in the dataset.

## .columns is used to give the column names in the dataset
coll_data.columns
```

```python
#3) Calculate the mean of each column of the dataset

## The mean of the 'SL' column in the dataset ##
coll_data['SL'].mean()
```

```python
## The mean of the 'SW' column in the dataset ##
coll_data['SW'].mean()
```

```python
## The mean of the 'PL' column in the dataset ##
coll_data['PL'].mean()
```

```python
#4) Check for the null values present in the dataset.

## use both .isna() and .sum() to find null values
### if value is '1' there is null_value(NaN) present, if value is '0' no null_value(NaN) present
coll_data.isna().sum()
```

```python
#5)Perform meaningful visualizations using the dataset.
import matplotlib.pyplot as plt
import seaborn as sns
# figure(figsize) helps to  adjust the size
plt.figure(figsize=(6,6))

# syntax for plotting histogram with 0.9 spacing 
plt.hist(col_data['SL'],rwidth=0.9)

# used to give the title of the graph
plt.title('Distribution of SL',fontsize =16)

# used to label the x axis and y axis
plt.xlabel('SL')
plt.ylabel('Frequency')

# used to denote the positions where ticks will be displayed
plt.xticks()
plt.yticks()
```

```python
# figure(figsize) helps to  adjust the size
plt.figure(figsize=(6,6))

# syntax for plotting histogram with 0.9 spacing
plt.hist(col_data['PL'],rwidth=0.9)

# used to give the title of the graph
plt.title('Distribution of PL',fontsize =16)

# used to label the x axis and y axis
plt.xlabel('PL')
plt.ylabel('Frequency')

# used to denote the positions where ticks will be displayed
plt.xticks()
plt.yticks()
```

```python
# figure(figsize) helps to  adjust the size
plt.figure(figsize=(6,6))

# syntax for plotting scatter diagram with size 10
plt.scatter(col_data['SL'],col_data['PL'],s=10)

# used to give the title of the graph
plt.title('Distribution of SL vs PL',fontsize =16)

# used to label the x axis and y axis
plt.xlabel('SL')
plt.ylabel('PL')

# used to denote the positions where ticks will be displayed
plt.xticks()
plt.yticks
```

```python
# get distribution of every data of each columns using pairplot

## figure(figsize) helps to  adjust the size
plt.figure(figsize=(10,10))

# syntax for calling pairplot function using seaborn
sns.pairplot(coll_data)

# Tight_layout keeps graphs from overlapping 
plt.tight_layout()
```

```python
# syntax used to plot two or more distribution at the same time
coll_data[['PL','PW']].plot.hist()

# used to give grid lines
plt.grid()
```
