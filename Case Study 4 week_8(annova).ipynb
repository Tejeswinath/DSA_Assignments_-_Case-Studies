# 2) Total Sales Generated for All Zones Each Month

```python
Total_Sales=sal_dat.groupby(['Zone - A','Zone - B','Zone - C','Zone - D']).sum().sort_values(by='Zone - D', ascending=False)
Total_Sales
```

# 3) Check for the Sales Amount in All the Zones

```python
#H0: All zones generate the same amount of sales.
#H1: All zones do not generate the same amount of sales.

from scipy.stats import stats
F,p=stats.f_oneway(sal_dat['Zone - A'],sal_dat['Zone - B'],sal_dat['Zone - C'],sal_dat['Zone - D'])
print("p_value for significance is: ",p)

if p<0.05:
    print("\nThere is ample evidence to reject null hypothesis")
else:
    print("There is ample evidence for failing to reject null hypothesis")
```
